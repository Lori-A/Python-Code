```python
#Lori-Ann Nembhard
# This function utilizes 3 functions and depending on the user's choice, it will display the corresponding function

def rock_paper_scissors():
    """
    This is a function that will ask the user to play rock, paper, scissors
    
    """
    import random

    print("Welcome to my game. Enjoy !")

    player = input("Do you want to be 'rock', 'paper', 'scissors'? ")
    print("You are", player)

    computer =  random.choice(['rock', 'paper', 'scissors'])
    print("The computer has", computer)


    if computer == player:
        print("It's a tie")
    elif player == 'rock':
        if computer == 'scissors':
            print("You lose!")
        else:
            print("You WIN!!!")
    elif player == 'paper':
        if computer == 'scissors':
            print("You lose!")
        else:
            print("You WIN!!!")
    elif player == 'scissors':
        if computer == 'paper':
            print("You WIN")
        else:
            print("You lose!")
    elif player == 'scissors':
        if computer == 'rock':
            print("You WIN!!!")
        else:
            print("You lose!")

def dating_game():
    """
    This is a function that will determine whether the user can date someone older or younger than their age
    """
    your_age = int(input("Enter your age: "))
    love_age = int(input("Enter the age of the person you want to date: "))
    movie = input("Do you like movies? Yes or No: ")

    if love_age < your_age / 2 + 7:
        print("You are a creeper!")
        if movie == "no":
            print("Uh oh! Get away from me!")
        else:
            print("You love movies but you're still a creeper!")
        
    elif love_age > (your_age - 7) * 2:
        print("You are good to go!")
        if movie == "yes":
            print("We both love movies!")
        else:
            print("I'm sorry. Not intersted. You dont like movies")
            
def turtle_artwork():
    """
    This is a function that will display a venn diagram using turtle.
    
    """
    import turtle
    turtle.showturtle()
    turtle.pencolor("pink")
    turtle.circle(100)
    turtle.forward(5)
    turtle.pencolor("violet")
    turtle.circle(100)
    turtle.penup()
    turtle.goto(120,0)
    turtle.pendown()
    turtle.pencolor("turquoise")
    turtle.circle(100)
    turtle.forward(5)
    turtle.circle(100)
    turtle.pencolor("black")
    turtle.penup()
    turtle.goto(70,-110)
    turtle.pendown()
    turtle.pencolor("blue")
    turtle.circle(100)
    turtle.forward(5)
    turtle.pencolor("purple")
    turtle.circle(100)



def main():
    """
    This is the main for the functions listed above
    
    """
    userChoice = 'Y'
    
    while userChoice == 'Y':
        
        print("\nWelcome! Lets get started!")
        print("Choose from the following choices:\n\n1. Rock-Paper-Scissors \n2. Dating Game \n3. Turtle Art")
        userInput= input("\nChoose R for Rock-Paper-Scissors, D for Dating Game, T for Turtle Art: ")

        if userInput == 'R' :
            rock_paper_scissors()
        elif userInput == 'D':
            dating_game()
        else:
            turtle_artwork()
        userChoice = input("\nDo you want to play again Y/N? ")
    print("Bye")
    
    
    
if __name__ == "__main__":
    main()
```

```python
#Lori-Ann Nembhard
#This program utilizes three functions. The first returns the cube of a number, the second function returns whichever
#number is closer to 44. The third function returns the sum of two numbers and displays it in the main

def cube(x):
    """
    This function will take cube a numeric value entered
    
    """
    return x**3


def closer_to_44(x,y):
    """
    This function will receive two values and determine which one is closer to 44
    """
    if abs(x-44) > abs(y-44):
        message = str(y) + " is closer to 44"
      
    elif abs(x-44) < abs(y-44):
        message = str(x) + " is closer to 44"
    else:
        message = "They are equal"
    return message
    
    
def sum_dice_roll():
    """
    This function will sum the number of two random numbers
    """
    import random
    
    die1 = random.randint(1,6)
    die2 = random.randint(1,6)
    return die1 + die2

    
def main():
    """
    This is the main for the functions listed above
    
    """
    userChoice = 'y'
    
    while userChoice == 'y':
        print("Enter: \n1 - Cube Function \n2 - Closer to a number \n3 - Two Die Function\n")
        choice = int(input("What is your choice? "))
        
        if choice == 1:
            
            print("Cube Function\n")
            number = int(input("Enter a number:"))
            print("The result is ", cube(number))
         
    
        elif choice == 2:
            print("Close to a number function\n")
            num1 =int(input("Enter a number: "))
            num2 = int(input("Enter a another number: "))
            print(closer_to_44(num1,num2))
    
    
        else:
            print("Two Die Function\n")
            print("The sum of two random numbers is", sum_dice_roll())
            
        
            
        userChoice = input("\nDo you want to play again y/n? ")
    print("Bye")

    
    
if __name__ == "__main__":
    main()
```
