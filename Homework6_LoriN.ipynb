```python
# Author: Murach and Yang
# Modified by: Lori-Ann Nembhard
# Movie - Phase 2 - Movie Name and Movie Year
# This program is designed to add movies along with the year and price to a two dimensional list which can later
# be retrieved or deleted.


def display_menu():
    """
    This function displays a list of commands that users will choose from
    
    """
    print("COMMAND MENU")
    print("list - List all movies")
    print("add - Add a movie")
    print("del - Delete a movie")
    print("exit - exit program")
    print("find - find year movie was released")
    
def list(movie_list):
    """
    This function displays a list of movies along with the year they were released and the price
    
    """
    if len(movie_list) == 0:
        print("There are no movies in the list.\n")
    else:
        i = 1
        print()
        for row in movie_list:
            print(str(i) + ". " + row[0] + " (" + str(row[1]) + ")"+ "- $"+ str(row[2]))
            i += 1
        
        print()
    
def add(movie_list):
    """
    This function allows the user to add a  movie to the list along with the year and price
    
    """
    name = input("Movie Name: ")
    year = int(input("Movie Year: "))
    ticket_price = int(input("Enter ticket price: "))
    movie = []
    movie.append(name)
    movie.append(year)
    movie.append(ticket_price)
    movie_list.append(movie)
    print(movie[0] + " was added.\n")

def find_by_year(movie_list):
    """
    This function allows the user to find movies based on the year they were released
    """
    movieYear = int(input("Enter movie year: "))
    i = 1 
    for movie in movie_list:
        if movie[1] == movieYear :
            print(movie[0] + " was released in " + str(movieYear))
            
            i+=1
    print()
            
def delete(movie_list):
    """
    This function allows the user to delete a movie from the list 
    
    """
    number = int(input("Movie number: "))
    if number < 1 or number > len(movie_list):
        print("Invalid movie number.\n")
    else:
        movie = movie_list.pop(number-1)
        print(movie[0] + " was deleted.\n")
              
def main():
    """
    This is the main for the functions listed above
    
    """
    movie_list = [["A Star is Born", 2018, 6.99],
                     ["An Old Man and a Gun", 2018, 5.99],
                    ["The Spy Who Dumped Me", 2017, 8.99]]
              
    display_menu()
    command = input("Enter Command: ")
        
    while (command.lower() != "exit"):
        
        if command.lower() == "list":
            list(movie_list)
            display_menu()
        elif command.lower() == "add":
            add(movie_list)
            display_menu()
        elif command.lower() == "del":
            delete(movie_list)
            display_menu()
        elif command.lower() == "find":
            find_by_year(movie_list)
            display_menu()
        else:
            print("Not a valid command.  Please try again.\n")

        command = input("Enter Command: ")
         
    print("You'll be back!")     
        
if __name__ == "__main__":
    main()
  
          
```
