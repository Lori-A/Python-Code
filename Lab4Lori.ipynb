### Lab 4, problem 2

```python
#Lori-Ann Nembhard
#This program utilizes a for loop with a range of values that add even and odd numbers as well as the number themselves

sumOfIntegers = 0
sumOfEvenNumbers = 0
sumOfOddNumbers = 0

for num in range(0,101):
    if num % 2 == 1:
        sumOfOddNumbers += num
    else:
        sumOfEvenNumbers += num
        
sumOfIntegers = sumOfOddNumbers + sumOfEvenNumbers

print("The sum of all even numbers between 1 and 100:", sumOfEvenNumbers, "\n")
print("The sum of all odd integers between 1 and 100:", sumOfOddNumbers,"\n")
print("The sum of all integers between 1 and 100:",sumOfIntegers,"\n" )
```

```python
#Lori-Ann Nembhard
#This program will ask the user enter their grade and by the end displays the average of all the grades entered.

count = 0
testScores = 0

userInput= int(input("Enter a test score between 0 and 100, -1 when done: "))


while userInput != -1:
   
    if userInput >= 0 and userInput <= 100:
        testScores += userInput
        count+= 1
    else:
        print("Invalid entry!")
    userInput= int(input("Enter a test score between 0 and 100, -1 when done: "))
    average = testScores/count
    average = round(average,2)
    
print("\nThe average of ", count, "test scores is:", average)
        
```

```python
#Lori-Ann Nembhard
# This program will display a coin toss and depending on which side it lands on (head for 1 and 2 for tails), 
# it will count the number of wins and add to your wallet and count the number of losses and deduct from your wallet

import random
numberOfWins = 0
numberOfLosses = 0
dollar = 500.0

for i in range(100):
    coinFlip = random.randint(1,2)

    if coinFlip == 1:
        print("The coin landed on heads.\nYou WIN!")
        numberOfWins += 1
        dollar += 10
    
    else:
        print("The coin landed on tails. \nYou lOSE!")
        numberOfLosses += 1
        dollar -= 10

print("\nNumber of heads: "+ str(numberOfWins) + "\nNumber of Tails: "+ str(numberOfLosses))
print("You now have $"+ str(dollar)+" in your wallet.")
```
